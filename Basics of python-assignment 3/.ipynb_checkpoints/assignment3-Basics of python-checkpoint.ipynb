# Exercise 1: Prime Numbers
Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.


```python
def PrimeCheck(n):
    if n <= 1:
        return False
    for i in range(2,int(n**0.5)+1): # Taking the range from 2 to sqrt(n) so we narrow down the seaarch
        if n % i == 0:   #checking if it is a prime or not based on divisibility
            return False
    else:
        return True
    
PrimeCheck(7) #checking the input num is prime or not
```

# Exercise 2: Product of Random Numbers
Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


```python
import random 

#generating random numbers in the range of 1 to 100
num1= random.randint(1,100) 
num2 = random.randint(1,100)
prod = num1 * num2

try:
    
    user_answer = int(input(f'Enter the product of {num1} and {num2}: '))

    if user_answer == prod:
        print(f'The product {user_answer} of num1: {num1} and num2: {num2} is correct')
    else:
        print(f'The product {user_answer} of num1: {num1} and num2: {num2} is incorrect')
        
except ValueError:   # using value error for invalid inputs
    print('The Number is invalid! Please enter a appropriate number!')
```

# Exercise 3: Squares of Even/Odd Numbers
Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.


```python
for j in range(100,201, 2): # we are taking even numbers from 100 with step 2
    squares = j**2
    print(f'The square of {j}  is:', squares)
```

# Exercise 4: Word counter
write a program to count the number of words in a given text.
example:
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
Expected output:
'This': 2 
'is': 1
'a': 1
'sample': 1
'text.': 1


```python
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
words = input_text.split() # splitting the text to take only words
word_count = {}   # creating an empty dictionary for storing count with words
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
        
print(word_count)
    
```

# Exercise 5: Check for Palindrome
Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.
Example:
Input: "racecar"
Expected Output: True


```python
def is_palindrome(str1):
    cleaned_text = ''
    
    for char in str1:
        if char.isalnum(): #checks if all the characters in a string are alphanumeric. if spaces and pnctuations exists then loop wont run.
            cleaned_text += char.lower()  # we lower the text for processing
            
    return cleaned_text == cleaned_text[::-1] # and reverse it to check if matches the original

print(is_palindrome('madam'))
print(is_palindrome('racecar'))
print(is_palindrome('hello world'))
```
