# Background:
Bombay hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. The operating cost for a franchise in a week (W) is given by the equation W = $1,000 + $5X, where X represents the number of units produced in a week. Recent feedback from restaurant owners suggests that this cost model may no longer be accurate, as their observed weekly operating costs are higher.

# Objective:
To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.
Data Provided:

•	The theoretical weekly operating cost model: W = $1,000 + $5X

•	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050

•	Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units


# hypothesis

H0 - there is no change in weakly operating costs. the model is accurate.

H1 - there is a significant change in weakly operating costs. the actual costs are higher than the models is currently suggesting.

we are given a problem where a  companies cost model is predicting operating costs wrong. 
     W = 1000 + 5X
   where , X - number of units produced per week
   
   given,
       sample size = 25
       sample mean =3050
       mean units produced (mean * x)= 600
       std_dev units(sigma * x)= 25
       
       so basically the theorotical mean under Null Hypothesis is:
           1000+ 5(600) = 4000

```python
n = 25  # sample size
sample_mean = 3050 # mean from 25 restraurants
std_dev = 5*25 #std_dev of units produced
mean_units = 600

theorotical_mean= 1000+ (5 * mean_units)
theorotical_std_dev = 5 * n


print(f'theorotical_mean: {theorotical_mean}')
print(f'theorotical_std_dev: {theorotical_std_dev}')
```

# test stats

because the population standard dev is known  we will use Z -test

The z-test formula is:
z = (x - mu) / (sigma / sqrt(n))
Where:
x is the sample mean,
mu is the population mean,
sigma is the population standard deviation,
n is the sample size

```python
z = (sample_mean - theorotical_mean) / (theorotical_std_dev / (n ** 0.5))
print(z)
```

# determining critical value
alpha = 0.05 

costs are higher so we must look for right tailed test in z table. which is approx 1.645

```python
critical_value = 1.645

if z > critical_value:
    print('Reject theh Null Hypothesis')
else:
    print('fails to reject the null hypothesis')
```

# Coclusion

we failed to reject the null hypothesis. there is no evidence that supports the retraurant owners claim that the costs are higher.
